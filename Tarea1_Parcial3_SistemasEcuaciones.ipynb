```
##Ramos Hern√°ndez Antonio

import matplotlib.pyplot as plt


def x2(y):
    return  0.25 + (0.25*y)
def y2(x,z):
    return  0.25 +(0.25*x) + (0.25*z)
def z2(y,w):
    return  0.25 + (0.25*y) + (0.25*w)
def w2(z):
    return  0.25 + (0.25*z)


def g_next():
    
    
    x = [0]
    y = [0]
    z = [0]
    w = [0]
    i=0
     
    while (i<10):
        
            x.append(x2(y[i]))
            y.append(y2(x[i],z[i]))
            z.append(z2(y[i],w[i]))
            w.append(w2(z[i]))
            i+=1
            
    
    print('x: '+ str(x) )
    print('y: '+ str(y) )
    print('z: '+ str(z) )
    print('w: '+ str(w) )
    
    plt.plot(x,label="x",color='b')
    plt.plot(y,label="y",color='r')
    plt.plot(z,label="z",color='y')
    plt.plot(w,label="w",color='g')
    
    plt.title("Sistema de ecuaciones ", fontsize = 24)
    
    plt.legend()
    plt.show()

    

g_next()
```
